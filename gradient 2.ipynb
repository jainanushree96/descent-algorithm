```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
dx=pd.read_csv("linearX.csv")
dy=pd.read_csv("https://raw.githubusercontent.com/coding-blocks-archives/machine-learning-online-2018/master/Datasets/Linear%20Regression/linearY.csv")
```

```python
dx=dx.values
dy=dy.values
```

```python

dx=dx.reshape((-1,))
dy=dy.reshape((-1,))
```

```python
dx=(dx-dx.mean())/dx.std()
```

```python
plt.scatter(dx,dy)
```

```python
def error(x,y,theta):
    m=x.shape[0]
    err=0
    for i in range(m):
        err+=(theta[0]+theta[1]*x[i]-y[i])**2
    return err
    
```

```python
def gradient(x,y,theta):
    grad=[0,0]
    m=x.shape[0]
    for i in range(m):
        grad[0]+=(theta[0]+theta[1]*x[i]-y[i])
        grad[1]+=(theta[0]+theta[1]*x[i]-y[i])*x[i]
    return grad
    
```

```python
def gradientdescent(x,y,learning_rate=0.001):
    theta=[-2.0,0.0]
    itr=100
    error_l=[]
    theta_list=[]
    
    for i in range(itr):
        err=error(x,y,theta)
        error_l.append(err)
        grad=gradient(x,y,theta)
        theta_list.append([theta[0],theta[1]])
        theta[0]=theta[0]-learning_rate*grad[0]
        theta[1]=theta[1]-learning_rate*grad[1]
    return theta,error_l,theta_list
```

```python
theta,elist,tlist=gradientdescent(dx,dy)
tlist=np.array(tlist)
```
